```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generar datos para la distribución

media = 70 #Media
desv_std = 4 #Desviación esatándar
num_datos= 500

# Generar los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
datos[0:50] # Mostrar los prmeros 50
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='blue') # Histograma de los datos
```

```python
# Visualizar la PDF
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 100)
pdf = norm.pdf(x, loc=media, scale=desv_std)
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'. format(np.round(media, 2), np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en kg. de los estudiantes')
plt.ylabel('Densisdad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True,color='b')
plt.plot( x, pdf, 'k', linewidth=2,)

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

Ejercicio 2 Distribución Normal

Una revisión de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviación estándar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg.
Cuantos estudiantes pesan más de 68 y menos de 72 kg.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parametros de la distribución normal
media = 70 # media
desv_std = 4 # Desviación estándar
total_datos = 500 # Total de datos

Peso_limite = 72
prob_menos_72kg = norm.cdf(Peso_limite, media, desv_std) # Changed 'peso_limite' to 'Peso_limite'
# Cálcular el número de estudiantes que pesan menos de 75kg
estudiante_menos_72kg = total_datos * prob_menos_72kg # Changed 'total_estudiantes' to 'total_datos'

print("Probabilidad de que un estudiante pese menos de 72 kg:", prob_menos_72kg)
print("Número de estudiantes que pesan de 72 kg:", estudiante_menos_72kg)

# Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= Peso_limite), color='skyblue', alpha=0.5) # Changed 'peso_limite' to 'Peso_limite'
plt.title('Probabilidad de que un estudiante pese menos de 72 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
# Parametros de la distribución normal
media = 70 # media
desv_std = 4 # Desviación estándar
total_datos = 500 # Total de datos

Peso_limite = 68
prob_menos_68kg = norm.cdf(Peso_limite, media, desv_std)

estudiante_menos_68kg = total_datos * prob_menos_68kg
```

```python
# Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= Peso_limite), color='skyblue', alpha=0.5) # Changed 'peso_limite' to 'Peso_limite'
plt.title('Probabilidad de que un estudiante pese menos de 68 kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

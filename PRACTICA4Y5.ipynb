```python
# Basada en el LIBRO estadistica para administración pagina 12
# Creamos una lista vacia que reprensenta la edad de 100 personas
import random
edades = []
#Hacemos un ciclo para generar 100 numeros aleatorios que representan las edades de loas persoans
for i in range(100):
  # Generamos un nunero aleatorio con la funcion randint y lo guardamos en la variable
  # Numero_aleatorio
  numero_aleatorio = random.randint( 18, 70)
  #Agreagamos el numero aleatorio a la lista de bvalores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# Creamos un Dataframe de Pandas con los números aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# Calcular la tabla de dsitribución de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla de distribución de frecuencia
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia']. cumsum()
tabla_frecuencias
```

# Reglas de sturges
1. Es una regla que sirve para calcular el número de clases o intervalos idóneo en los que se debe dividir un conjunto de datos
2.


```python
# Importa la libreria math de funciones matematicas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
# Creea un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
# Calcular el número de intervalos utilizando la regla de Sturgera
n = len(edades)
k = int (1 + math.log2(n))
print("Valor de n =", n , " \nNumero de intervalos =", k)
```

```python
# Crear la tabla de frecuencia

tabla_frecuencias= pd.cut(df['Edad'], bins=k, include_lowest=True). value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
# Orden la tabla de frecuencia por intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
# Generar columnas siguientes
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Porc Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()

#Mostrar la tabla de frecuencia
print(tabla_frecuencias)
```

```python
import matplotlib.pyplot as plt
plt.hist(edades, alpha = 0.5,bins = 20)
plt.title("Histograma de edades basado en la funcion de matplotlib con valores estandar")
```

```python
# Crear un histograma basado en la tabla de frecuencia
plt.hist(df['Edad'], bins=k, alpha = 0.5)
plt.title('Histograma de edades basado en la regla de sturges')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```

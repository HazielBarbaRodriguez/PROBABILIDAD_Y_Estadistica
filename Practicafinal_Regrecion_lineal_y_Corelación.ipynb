```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("grupo1.csv")
df1.head()
```

```python
df2=pd.read_csv("grupo2.csv")
df2.head()
```

```python
df3=pd.read_csv("grupo3.csv")
df3.head()
```

```python
print("Primeras filas del DataFrame grupo1:")
print(df1.head())


print("\nDescripción estadística:")
print(df1.describe())

plt.figure(figsize=(8, 5))
sns.histplot(df1['Calificacion'], bins=10, kde=True, color='skyblue')
plt.title('Histograma de Calificaciones (Grupo 1)')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
print("Primeras filas del DataFrame grupo2:")
print(df2.head())

print("\nDescripción estadística del DataFrame grupo2:")
print(df2.describe())

plt.figure(figsize=(8, 5))
sns.histplot(df2['Calificacion'], bins=10, kde=True, color='orange')
plt.title('Histograma de Calificaciones (Grupo 2)')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
print("Primeras filas del DataFrame grupo3:")
print(df3.head())

print("\nDescripción estadística del DataFrame grupo3:")
print(df3.describe())

plt.figure(figsize=(8, 5))
sns.histplot(df3['Calificacion'], bins=10, kde=True, color='green')
plt.title('Histograma de Calificaciones (Grupo 3)')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
#Modelo de Regresión Lineal:
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)

X = df[['HorasEstudio']]  # Variable independiente
y = df['Calificacion']     # Variable dependiente

#Entrenar el modelo regresión líneal
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Visualizacion
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='HorasEstudio', y='Calificacion', color='blue', alpha=0.6)
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')
plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualizacion de Resultados
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df, color='blue', label='Datos reales', alpha=0.6)

y_pred = modelo.predict(X)


plt.plot(df['HorasEstudio'], y_pred, color='red', label='Línea de regresión')

plt.title('Regresión Lineal: Horas de Estudio vs Calificación')
plt.xlabel('HorasEstudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Cálculo de la correlación
from scipy.stats import pearsonr

# Asegúrate de que no haya valores nulos
df = df.dropna(subset=['HorasEstudio', 'Calificacion'])

# Calcular coeficiente de correlación de Pearson
coef_pearson, p_valor = pearsonr(df['HorasEstudio'], df['Calificacion'])

# Mostrar resultados
print(f" Coeficiente de correlación de Pearson: {coef_pearson:.4f}")
print(f" Valor p (significancia estadística): {p_valor:.4f}")
```
